<h1>Extraccion de Datos Covid 19</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv('./owid-covid-data.csv')
```

```python
df.tail(5)
```

```python
df = df[ df.date == '2021-11-12' ]
```

```python
df.head(10)
```

```python
df = df.reset_index()
```

```python
# creando dataframe de continentes
continent_data = df[ df.continent.isna() ]
```

```python
continent_data.head()
```

```python
continent_data = continent_data.drop('continent', axis=1)
```

```python
continent_data[ (continent_data.location == 'Asia') | 
                (continent_data.location == 'Africa') |
                (continent_data.location == 'Oceania') |
                (continent_data.location == 'North America') |
                (continent_data.location == 'South America') |
                (continent_data.location == 'Europe') 
                 ]
```

```python
# eliminando datos nulos de la colummna continent
df = df.dropna( subset=['continent'] )
```

```python
df
```

```python
#exportando csv
df.to_csv('./covid.csv', index=False)
```
